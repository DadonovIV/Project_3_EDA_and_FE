# Основы Kaggle

    Kaggle — международная платформа для проведения соревнований по анализу данных на основе реальных задач от различных компаний. Эта площадка может похвастаться самым большим Data Science-комьюнити со всего мира — более 200 тысяч участников, с которыми можно взаимодействовать посредством разных активностей.

Kaggle отлично помогает как начинающим, так и продвинутым дата-сайентистам развивать свои навыки: не только в решении прикладных задач, но и в работе с данными, визуализации, генерации идей и признаков, выборе подходящих моделей и так далее.

Площадка Kaggle была основана в 2010 году Энтони Гольдблюмом и Джереми Ховардом. В июне 2017 года количество зарегистрированных пользователей площадки превысило 1 млн человек. В этом же году площадку приобретает Google.

В 2021 году на Kaggle насчитывается уже 8 млн пользователей из 194 стран. 

Далее

* узнаем, как зарегистрироваться на Kaggle и пройти верификацию номера телефона (это важный шаг для дальнейшей работы на площадке!);
* обновим описание своего профиля и добавим нескольких друзей;
* разберёмся с уровнями мастерства, рейтингами и медалями.

Чтобы зарегистрировать аккаунт, необходимо перейти на главную страницу Kaggle и в правом верхнем углу или в левой части страницы нажать на кнопку Register. После указания почты, пароля и имени пользователя на указанный адрес почты будет выслан шестизначный код, который нужно будет ввести на следующем шаге. 

Чтобы посмотреть детали своего профиля, статистику или внести изменения в информацию, нужно нажать на иконку профиля в правом верхнем углу и перейти по ссылке My Profile. 

Чтобы заполнить профиль, необходимо перейти во вкладку Account, нажать на кнопку Edit Public Profile, где можно указать род своих занятий, компанию, город нахождения (например, DS at Home, Moscow). 

На главной странице профиля (Home) также можно заполнить поле Bio. 

Чтобы пользоваться всеми преимуществами площадки, необходимо пройти верификацию своего профиля, добавив номер мобильного телефона. 

Если вы столкнулись с проблемами при верификации, в левой навигационной панели нажмите на кнопку More → Support/Contact → I cannot activate my account (первая ссылка сверху) → I cannot verify my phone number и отправьте сообщение поддержке через форму.  

## РАНГИ ПОЛЬЗОВАТЕЛЕЙ KAGGLE

Площадка Kaggle важна для становления дата-сайентиста: работодатели оценивают активность кандидатов и их достижения на площадке. Для обоих участников рынка труда это удобный инструмент: работодатели имеют возможность увидеть, как кандидат пишет код, излагает мысли, что загружает; а работник может использовать площадку как своё портфолио.

Если перейти во вкладку Competitions, можно увидеть список соревнований разных видов. 

* Соревнование типа Knowledge. За этот вид соревнований для новичков баллы не начисляются. 
* Соревнования с денежной наградой. 
* Соревнования, привязанные к определённым курсам, мероприятиям и так далее. 

Участие в соревнованиях влияет на публичный рейтинг пользователя. 

Чтобы увидеть общую таблицу рейтинга, нужно развернуть разделы, нажав на кнопку More → User Rankings. 

Рейтинги составляются по четырём видам активностей: 

1. Competitions (участие в соревнованиях);
2. Datasets (публикация открытых датасетов);
3. Notebooks (создание ноутбуков);
4. Discussions (публикация сообщений на форуме).

Пользователи Kaggle могут последовательно достичь следующих пяти разных уровней статуса в любой из категорий: Novice, Contributor, Expert, Master и Grandmaster. 

    ❗️ Для достижения каждого из этих уровней требуется выполнить определённые условия, которые усложняются с каждым шагом в сторону Grandmaster. Рейтинги Kaggle важны для карьеры любого дата-сайентиста: на текущий момент почти все работодатели обращают пристальное внимание на успехи кандидата на этой площадке. Как правило, работодатели при просмотре профиля соискателя на Kaggle обращают внимание на уровни в категориях Competitions и Notebooks — именно эти категории позволяют оценить специфические навыки дата-сайентиста.

1. Novice (Новичок). Для получения ранга достаточно зарегистрироваться на платформе.
2. Contributor (Активный пользователь). Для получения ранга необходимо запустить один ноутбук или скрипт + подать одну заявку на соревнование + оставить один комментарий + один раз проголосовать. 
3. Expert (Эксперт). Для получения ранга необходимо 2 бронзовых медали в соревнованиях ИЛИ 3 бронзовых медали за датасеты ИЛИ 5 бронзовых медалей за ноутбуки ИЛИ 50 бронзовых медалей за обсуждения.
4. Master (Мастер). Для получения ранга необходима 1 золотая медаль + 2 серебряные медали в соревнованиях ИЛИ 1 золотая медаль + 4 серебряных медали за датасеты ИЛИ 10 серебряных медалей за ноутбуки ИЛИ 50 серебряных медалей + 200 медалей в общем за обсуждения.
5. Grandmaster (Грандмастер). Для получения ранга необходимо 5 золотых медалей + одна сольная золотая медаль в соревнованиях ИЛИ 5 золотых медалей + 5 серебряных медалей за датасеты ИЛИ 15 золотых медалей за ноутбуки ИЛИ 50 золотых медалей + 500 медалей итого за обсуждения.

Медали даются за достижения и активность в каждой категории (соревнования, датасеты, ноутбуки, обсуждения). На Kaggle разработана достаточно комплексная система выдачи медалей в зависимости от разных факторов. 

Например, бронзовая медаль в соревнованиях даётся за попадание в ТОП-40 % результатов в команде до 99 игроков и в ТОП-10 % результатов в команде из более чем 1000 игроков.

## СОРЕВНОВАНИЯ НА KAGGLE

Откроем уже завершённое соревнование на Kaggle и на его примере познакомимся с этим важнейшим элементом площадки.  

На странице каждого соревнования вы увидите несколько вкладок: 

* **Overview.** Во вкладке расположена информация с описанием соревнования (Description) и об оценке решения (Evaluation). Обращайте особое внимание на эту информацию: в дальнейшем вы убедитесь, что внимательное отношение к метрикам может значительно изменить итоговую оценку вашего решения!

        Обратите внимание на блок информации в нижней части раздела Description. Здесь вы увидите данные об участниках и о том, учитываются ли результаты соревнования в обновлении рангов и начисляются ли за него очки рейтинга.

![](./dst-eda-7-39.png)

* **Data.** Здесь располагаются датасеты к соревнованию. Скачать файлы можно только после согласия с правилами соревнования.
* **Code.** Здесь находятся ноутбуки и бейзлайны, созданные как вами, так и другими участниками соревнования. Правило хорошего тона — ставить Upvote ноутбукам.
* **Discussion.** В этом разделе можно принять участие в обсуждении, поблагодарить авторов, задать вопрос и ознакомиться с комментариями сообщества.
* **Leaderboard.** В этом разделе отображаются результаты сабмитов — публичные и приватные. Очки и медали начисляются за результаты в приватных лидербордах. Обратите внимание на правила расчёта позиций в лидербордах. В этом соревновании позиции в публичном лидерборде рассчитываются по Score, полученному на 50 % тестовой выборки. Поскольку соревнование ещё не завершено, финальный результат будет рассчитан по Score, полученному на оставшейся половине тестовой выборки.


![](./dst-eda-7-40.png)

* **Rules.** Информация о правилах соревнования: можно ли объединяться в команды, какие библиотеки можно использовать и так далее.

## Работа с ноутбуками в Kaggle

В этом юните мы узнаем:

* для чего нужны ноутбуки на Kaggle и как их правильно настраивать для конкретной задачи;
* как сохранить результаты своей работы в ядре Kaggle;
* как добавить участников к своему ноутбуку и изменить настройки приватности.

Кроме того, мы выполним следующие действия: 

✔️ Создадим новый ноутбук с нуля для выбранного соревнования, подготовим файл сабмита и отправим его в лидерборд.

✔️ Склонируем ноутбук другого участника, обновим его и отправим результаты в соревнование.

✔️ Подключимся к соревнованию, над которым будем работать на протяжении модуля.

Чтобы присоединиться к соревнованию, достаточно нажать кнопку Join Competition и согласиться с правилами. 

→ Во вкладке Overview вы увидите описание задачи и условий соревнования (раздел Description) и информацию о метриках качества решения (раздел Evaluation). 

![](./dst-eda-7-7.png)

Наше соревнование посвящено предсказанию стоимости автомобиля по его характеристикам, а метрикой качества является MAE (Mean Absolute Error, среднее абсолютное отклонение).

Результаты этого соревнования не будут учитываться в рангах по соревнованиям, но засчитаются по обсуждениям, датасетам и ноутбукам. 

→  Вкладка Data содержит в себе информацию о датасетах. В нашем соревновании их три: 

* sample_submission.csv;
* test_data.csv (данные, на которых мы будем делать предсказание. Не содержат в себе целевую переменную final_price);
* train_data.csv (обучающая выборка).

→  Во вкладке Code находятся все ноутбуки, которые встречаются в соревновании. В нашем случае это ноутбук с бейзлайном и ноутбук, который нужно будет склонировать для дальнейшей работы. 

→  Во вкладке Leaderboard размещена информация о результатах участников соревнования.

## КАК СОХРАНИТЬ НОУТБУК?

Для сохранения ноутбука предназначена кнопка Save Version в правом верхнем углу. 

![](./dst-eda-7-14.png)

Можно создать фиксированную версию под конкретным названием, например Starter. К этой версии впоследствии можно будет вернуться.

Есть два варианта сохранения версии: 

* с запуском всех ячеек и сохранением вывода (Save & Run All);
* быстрое сохранение без запуска (Quick Save). 

В первом случае все ячейки ноутбука будут выполняться последовательно до конца или до первой ошибки. Во втором варианте сохранение производится без запуска в том виде, в каком вы написали ноутбук. 

![](./dst-eda-7-15.png)

В расширенных настройках (Advanced Settings) можно найти опции для сохранения вывода при быстром сохранении (с опциями: никогда не сохранять, всегда сохранять, сохранить вывод для конкретной версии). Рекомендуем выбрать опцию «Всегда сохранять» (Always save output). Также здесь можно использовать сохранение и запуск с ускорением (доступно при верифицированном номере телефона). 

![](./dst-eda-7-16.png)

После сохранения версии в нижнем левом углу вы увидите окно с перечислением активных событий: 

![](./dst-eda-7-17.png)

![](./dst-eda-7-18.png)

Перейдя в менеджер просмотров, вы увидите свой ноутбук. Здесь вы можете поставить себе Upvote. При нажатии на три точки слева вы увидите дополнительные действия, которые можно произвести с ноутбуком, например загрузить его на локальный диск.

![](./dst-eda-7-20.png)

## КЛОНИРОВАНИЕ НОУТБУКА И РАБОТА В НОУТБУКАХ ДРУГИХ УЧАСТНИКОВ

Помимо создания ноутбука с нуля, есть и вариант использования ноутбука, созданного другим участником. 

1. Перейдём во вкладку Code на странице соревнования. 
2. В ней, как мы помним, лежит бейзлайн-решение и ноутбук для клонирования (Clone my notebook). 
3. Откроем его, не забыв поставить Upvote (это хорошая практика). 
4. В верхней части экрана мы увидим кнопку Copy & Edit. 

![](./dst-eda-7-28.png)

После нажатия на эту кнопку мы перейдём в привычное окно редактирования ноутбука. Можно переименовать его, изучить содержимое и спокойно вносить свои изменения. 

Мы видим, что в ноутбуке импортируются стандартные библиотеки Pandas и NumPy, файлы соревнования, а в качестве сабмита взято среднее значение по выборке. 

Далее мы можем сохранить эту версию ноутбука, не забыв указать сохранение вывода (Always save output). 

При клонировании ноутбука остаётся информация о ноутбуке-источнике в верхней информационной части страницы.

## КАК ОТСЛЕЖИВАТЬ SCORE СВОИХ РЕШЕНИЙ?

Допустим, для сабмита вы решите использовать вместо среднего (mean) медиану (median). 

Если вы совершите Quick Save с настройкой Always save output, при просмотре ноутбука вы увидите два сабмит-файла: mean_submit и median_submit. 

![](./dst-eda-7-30.png)

На соревнование можно послать любой из существующих сабмитов. 

Как видите, использование медианы существенно улучшило Score.

![](./dst-eda-7-31.png)
